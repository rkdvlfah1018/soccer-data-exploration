```python
import pandas as pd
from statsbombpy import sb
from mplsoccer import Pitch
```

```python
#find all competitions
sb.competitions()
```

```python
#using competition_id 43 season_id 106 which is 2022 world cup
sb.matches(competition_id = 43, season_id = 106)
```

```python
#world cup final
MATCH_ID = 3869685
```

```python
#pull all the events in the match except for statsbomb_360 events
match_events_df = sb.events(match_id = MATCH_ID)

#get 360 data (360 data should be in local file)
match_360_df = pd.read_json(f'../open-data/data/three-sixty/{MATCH_ID}.json')
```

```python
#includes data like kick-off, subs, etc..
match_events_df['id']
```

```python
#only has the 360 data (freeze frame)
match_360_df['event_uuid']
```

```python
#to look into one row of df
match_360_df.iloc[0]
```

```python
#look into freeze_frame property
match_360_df.iloc[0]['freeze_frame']
```

```python
#merging the dataframes in a way that when the event data is not considering "playing ball" you add nothing
#but when it is, add 360 data to the row
df = pd.merge(left = match_events_df, right = match_360_df, left_on = 'id', right_on = 'event_uuid', how = 'left')
```

```python
df.head(25)
```

```python
df.columns
```

```python
#checking out Messi's first pass
MESSI = 5503 #this is Messi's player id
df = df[(df['player_id'] == MESSI) & (df['type'] == 'Pass')].reset_index(drop = True)
```

```python
#now have all of Messi's pass
df.head()
```

```python
#create 2 new columns x_start and y_start using the location column of the df
df[['x_start', 'y_start']] = pd.DataFrame(df.location.tolist(), index = df.index)
#same but now do the end location
df[['x_end', 'y_end']] = pd.DataFrame(df.pass_end_location.tolist(), index = df.index)
```

```python
#use mplsoccer package to visiualize the pass(es)
p = Pitch(pitch_type = 'statsbomb')
fig, ax = p.draw(figsize = (12,8))

df = df[0:1]

p.scatter(x=df['x_start'], y=df['y_start'], ax=ax)
p.lines(xstart=df['x_start'], ystart=df['y_start'], xend=df['x_end'], yend=df['y_end'], ax=ax, comet = True)

for x in df.iloc[0]['freeze_frame']:
    if x['teammate']:
        color = 'blue' #argentina
    else:
        color = 'red'
    
    p.scatter(x=x['location'][0], y=x['location'][1], ax=ax, c=color, s=100)
```
